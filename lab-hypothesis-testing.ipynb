* Challenge 1

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass.  
 Choose the propper test and, with 5% significance, comment your findings.

```python
# H0:  HP Dragons = HP Grass  (hipotesis nula)
# H1:  HP Dragons > HP Grass  (lo que queremos demostrar)

from scipy.stats import ttest_ind

dragon_hp = df[df['Type 1'] == 'Dragon']['HP'].dropna()  # miramos los Type_1 sus HP
grass_hp  = df[df['Type 1'] == 'Grass']['HP'].dropna()

t_stat, p_value = ttest_ind(dragon_hp, grass_hp, equal_var=False)

alpha = 0.05

print("p_value =", p_value)

if p_value < alpha:
    print("Rechazamos hipotesis nula  H0: Pokémon tipo Dragon tiene mas HP que Pokémon tipo Grass.")
else:
    print("No rechazamos la hipotesis nula , pero tampoco podemos asegurar que  Pokémon tipo Dragon tiene mas HP que Pokémon tipo Grass.")
```

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary.  
 Choose the propper test and, with 5% significance, comment your findings.


```python
df.head()
```

```python
# H0:  Legendary(Attack, Defense, Sp. Atk, Sp. Def, Speed)  = No_Legendary(Attack, Defense, Sp. Atk, Sp. Def, Speed)  (hipotesis nula)
# H1:  Legendary(Attack, Defense, Sp. Atk, Sp. Def, Speed)  != No_Legendary(Attack, Defense, Sp. Atk, Sp. Def, Speed)  (queremos demostrar)

from scipy.stats import ttest_ind

stats = ['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']
alpha = 0.05 

for col in stats:
    legendary = df[df['Legendary'] == True][col].dropna()   # Cada grupo limpia sus propios NaN, porque no depende del otro.
    non_legendary = df[df['Legendary'] == False][col].dropna()
    
    t_stat, p_value = ttest_ind(legendary, non_legendary, equal_var=False)


    print("p_value =", p_value )

    if round(p_value,2) < alpha:
        print(f"Rechazamos hipotesis nula {col} Legendary = {col} No_Legendary")
    else:
        print(f"No rechazamos la hipotesis nula , pero tampoco podemos asegurar que {col} Legendary = {col} No_Legendary")
    print("------------------------------------------")

   

    
```

Resultado :  
Utilizando una Two Sample T-test con un nivel de significancia del 5%, comparamos las estadísticas de Pokémon legendarios y no legendarios en HP, Ataque, Defensa, Sp. Atk, sp. Defensa y velocidad.
En los seis casos se rechaza la hipotesis nula para cada estadistica y concluimos que los Pokémon legendarios tienen estadísticas medias significativamente diferentes en comparación con los Pokémon no legendarios.

Vamos a ver la distancia

Funcion euclidea : para 2 puntos P1(x1,y1) p2(x2,y2) la distancia euclidea es:  
np.sqrt((x1 - x2)**2 + (y1 - y2)**2)

```python
import numpy as np

def dist(x1, y1, x2, y2):
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
School_coordinates= (-118, 34)
Hospital_coordinates = (-122, 37)

x1_school, x2_school = School_coordinates
x1_hospital, x2_hospital = Hospital_coordinates
```

```python
def dist(x1, y1, x2, y2):
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
df['distancia_hospital'] = dist(x1_hospital,x2_hospital,df['longitude'],df['latitude'])
df['distancia_school'] = dist(x1_school,x2_school,df['longitude'],df['latitude'])
df
```

A mi me gusta hacerlo con lambda que puedes poner lo que quieras en vez de la x con una condicion

```python
df["cercany_hospital"] = df["distancia_hospital"].apply(lambda x: True if x < 0.5 else False) # valor celda es 'lower' si x<0.5 sino 'far'
df["cercany_school"] = df["distancia_school"].apply(lambda x: True if x < 0.5 else False)
df['close_hospital_or_scholl'] = df["cercany_hospital"] | df["cercany_school"]
```

```python
df_estudio =df[['numero_casa', 'cercany_hospital', 'cercany_school','close_hospital_or_scholl','median_house_value']]
df_estudio
```

```python
close_hospital_or_scholl = df_estudio[df_estudio['close_hospital_or_scholl'] == True]['median_house_value']  # grupos cerca de alguno
far_hospital_or_scholl  = df_estudio[df_estudio['close_hospital_or_scholl'] == False]['median_house_value']   # grupos no cerca de ninguno
```

```python
# H0 =	Las casas cercanas y lejanas tienen el mismo precio medio  # Si p_value > alpha -> Aceptamos hipótesis nula
# H1 =  Las casas cercanas son más (>) caras que las lejanas ( la que ) # Si p_value < alpha -> Rechazamos hipótesis nula -> Aceptamos la hipótesis alternativa

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(close_hospital_or_scholl, far_hospital_or_scholl, equal_var=False) # evalua los dos grupos COMPARANDO CERCA O LEJOS EN RELACION AL PRECIO
print("t =", t_stat, "p =", p_value)
```

```python
alpha = 0.05

if p_value < alpha:
    print("RECHAZA hipotesis nula → Las casas mas cercanas son mas caras")
else:
    print("NO RECHAZA hipotesis nula pero No asegura que las casas sean mas caras por cercania")
```
